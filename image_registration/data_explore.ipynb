```python
import tifffile
from PIL import Image
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display
```

Open data file

```python
filename = "data/4-SIM-merge-18frames.tif"
data = tifffile.imread(filename)
```

```python
print(f"Dimensions of the image: {data.shape}")
print(f"Data type of the elements: {data.dtype}")
```

```python
def analyze_tiff_image(file_path):
    # Load the TIFF image
    with Image.open(file_path) as img:
        # Display basic information
        print(f"Format: {img.format}")
        print(f"Size: {img.size}")
        print(f"Mode: {img.mode}")

        # Display the image
        plt.imshow(img)
        plt.title("TIFF Image")
        plt.show()
```

```python
analyze_tiff_image(filename)
```

```python
# Function to update the plot
def update_plot(channel_index, slice_index):
    plt.imshow(data[channel_index, slice_index, :, :], cmap='gray')
    plt.show()

# Create interactive widgets
channel_slider = widgets.IntSlider(min=0, max=data.shape[0] - 1, step=1, description='Channel:')
slice_slider = widgets.IntSlider(min=0, max=data.shape[1] - 1, step=1, description='Slice:')

# Display the widgets and link the function
ui = widgets.VBox([channel_slider, slice_slider])
out = widgets.interactive_output(update_plot, {'channel_index': channel_slider, 'slice_index': slice_slider})

display(ui, out)
```
